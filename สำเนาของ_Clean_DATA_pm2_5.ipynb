```python
import pandas as pd
import numpy as np
```

```python
import glob

location = '/content/Datasci/*.xlsx'
excel_files = glob.glob(location)

DATA = pd.DataFrame()

for excel_file in excel_files:
    df2 = pd.read_excel(excel_file)
    DATA = pd.concat([DATA, df2], ignore_index=True)

DATA.fillna(value="N/A", inplace=True)
DATA.to_excel('/content/Datasci/comblined.xlsx')


print(DATA)
```

```python
df['05T'] = df['05T'].astype(float)
df['61T'] = df['61T'].astype(float)
df['54T'] = df['54T'].astype(float)
df['02T'] = df['02T'].astype(float)
df['59T '] = df['59T '].astype(float)

df.info()
```
