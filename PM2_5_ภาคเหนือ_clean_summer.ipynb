```python
summer_frames = [df.loc['2011-02-16':'2011-05-15'],
                 df.loc['2012-02-16':'2012-05-15'],
                 df.loc['2013-02-16':'2013-05-15'],
                 df.loc['2014-02-16':'2014-05-15'],
                 df.loc['2015-02-16':'2015-05-15'],
                 df.loc['2016-02-16':'2016-05-15'],
                 df.loc['2017-02-16':'2017-05-15'],
                 df.loc['2018-02-16':'2018-05-15'],
                 df.loc['2019-02-16':'2019-05-15'],
                 df.loc['2020-02-16':'2020-05-15'],]

df_summer = pd.concat(summer_frames)
df_summer
```

```python
df_summer.isnull().sum()
```

```python
df_summer[df_summer.isnull().all(axis=1)] 

#เริ่มมีข้อมูล2011-05-16 ถึง 2013-02-27(page4)
#ก่อน2011-05-16 = ลบทิ้ง
#2013ขาดช่วงเดือน3-4 = จะแทนที่ด้วยค่าเฉลี่ย
```

```python
df_summer11 = df_summer.loc['2011']
df_summer12 = df_summer.loc['2012']
df_summer13 = df_summer.loc['2013']
df_summer14 = df_summer.loc['2014']
df_summer15 = df_summer.loc['2015']
df_summer16 = df_summer.loc['2016']
df_summer17 = df_summer.loc['2017']
df_summer18 = df_summer.loc['2018']
df_summer19 = df_summer.loc['2019']
df_summer20 = df_summer.loc['2020']
```

```python
df_summer13.mean()
```

```python
df_summer11 = df_summer11.fillna(df_summer11.mean())
df_summer12 = df_summer12.fillna(df_summer12.mean())
df_summer13 = df_summer13.fillna(df_summer13.mean())
df_summer14 = df_summer14.fillna(df_summer14.mean())
df_summer15 = df_summer15.fillna(df_summer15.mean())
df_summer16 = df_summer16.fillna(df_summer16.mean())
df_summer17 = df_summer17.fillna(df_summer17.mean())
df_summer18 = df_summer18.fillna(df_summer18.mean())
df_summer19 = df_summer19.fillna(df_summer19.mean())
df_summer20 = df_summer20.fillna(df_summer20.mean())
```

```python
df_summer.loc['2011'] = df_summer11
df_summer.loc['2012'] = df_summer12
df_summer.loc['2013'] = df_summer13
df_summer.loc['2014'] = df_summer14
df_summer.loc['2015'] = df_summer15
df_summer.loc['2016'] = df_summer16
df_summer.loc['2017'] = df_summer17
df_summer.loc['2018'] = df_summer18
df_summer.loc['2019'] = df_summer19
df_summer.loc['2020'] = df_summer20

df_summer
```

```python
df_summer = df_summer[~(df_summer.index < '2011-05-16')]

df_summer
```

```python
df_summer['year'] = pd.DatetimeIndex(df_summer.index).year
df_summer
```

```python
year_view = pd.DataFrame(df_summer.groupby(by='year')[['57T','73T','35T','36T','67T','75T','70T','70T**','69T','58T','37T','38T','39T','40T','68T']]
                              .mean())
year_view
```

```python
year_sm = year_view 
year_sm['mean'] = year_view.mean(axis=1)
year_sm[['mean']]
```

```python
dfs = year_sm 
x = year_sm.index
y = year_sm[['mean']] 
```
