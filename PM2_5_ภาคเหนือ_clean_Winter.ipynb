1. ฤดูร้อน 
    16 ก.พ.(ปีนั้น) - 15 พ.ค.(ปีนั้น)
2. ฤดูฝน 
    16 พ.ค.(ปีนั้น) - 15 ต.ค.(ปีนั้น) 
3. ฤดูหนาว 
    16 ต.ค.(ปีนั้น) - 15 ก.พ. (ปีถัดไป)

```python
winter_frames = [df.loc['2011-10-16':'2012-02-15'],
                 df.loc['2012-10-16':'2013-02-15'],
                 df.loc['2013-10-16':'2014-02-15'],
                 df.loc['2014-10-16':'2015-02-15'],
                 df.loc['2015-10-16':'2016-02-15'],
                 df.loc['2016-10-16':'2017-02-15'],
                 df.loc['2017-10-16':'2018-02-15'],
                 df.loc['2018-10-16':'2019-02-15'],
                 df.loc['2019-10-16':'2020-02-15'],
                 df.loc['2020-10-16':'2020-12-31'],]

df_winter = pd.concat(winter_frames)
df_winter
```

```python
df_winter.isnull().sum()
```

```python
df_winter.isnull().sum()
```

```python
df_winter = df_winter.reset_index()
df_winter
```

```python
conditions = [
    (df_winter['Date'] >= '2011-10-16') & (df_winter['Date'] <= '2012-02-15'),
    (df_winter['Date'] >= '2012-10-16') & (df_winter['Date'] <= '2013-02-15'),
    (df_winter['Date'] >= '2013-10-16') & (df_winter['Date'] <= '2014-02-15'),
    (df_winter['Date'] >= '2014-10-16') & (df_winter['Date'] <= '2015-02-15'),
    (df_winter['Date'] >= '2015-10-16') & (df_winter['Date'] <= '2016-02-15'),
    (df_winter['Date'] >= '2016-10-16') & (df_winter['Date'] <= '2017-02-15'),
    (df_winter['Date'] >= '2017-10-16') & (df_winter['Date'] <= '2018-02-15'),
    (df_winter['Date'] >= '2018-10-16') & (df_winter['Date'] <= '2019-02-15'),
    (df_winter['Date'] >= '2019-10-16') & (df_winter['Date'] <= '2020-02-15'),
    (df_winter['Date'] >= '2020-10-16') & (df_winter['Date'] <= '2020-12-31')
    ]

values = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020'] 

df_winter['year'] = np.select(conditions, values)
df_winter
```

```python
df_wt11 = df_winter.loc[df_winter['year'].isin(['2011'])]
df_wt12 = df_winter.loc[df_winter['year'].isin(['2012'])]
df_wt13 = df_winter.loc[df_winter['year'].isin(['2013'])]
df_wt14 = df_winter.loc[df_winter['year'].isin(['2014'])]
df_wt15 = df_winter.loc[df_winter['year'].isin(['2015'])]
df_wt16 = df_winter.loc[df_winter['year'].isin(['2016'])]
df_wt17 = df_winter.loc[df_winter['year'].isin(['2017'])]
df_wt18 = df_winter.loc[df_winter['year'].isin(['2018'])]
df_wt19 = df_winter.loc[df_winter['year'].isin(['2019'])]
df_wt20 = df_winter.loc[df_winter['year'].isin(['2020'])]
```

```python
df_wt11 = df_wt11.fillna(df_wt11.mean())
df_wt12 = df_wt12.fillna(df_wt12.mean())
df_wt13 = df_wt13.fillna(df_wt13.mean())
df_wt14 = df_wt14.fillna(df_wt14.mean())
df_wt15 = df_wt15.fillna(df_wt15.mean())
df_wt16 = df_wt16.fillna(df_wt16.mean())
df_wt17 = df_wt17.fillna(df_wt17.mean())
df_wt18 = df_wt18.fillna(df_wt18.mean())
df_wt19 = df_wt19.fillna(df_wt19.mean())
df_wt20 = df_wt20.fillna(df_wt20.mean())
```

```python
df_winter.loc[df_winter['year'].isin(['2011'])] = df_wt11
df_winter.loc[df_winter['year'].isin(['2012'])] = df_wt12
df_winter.loc[df_winter['year'].isin(['2013'])] = df_wt13
df_winter.loc[df_winter['year'].isin(['2014'])] = df_wt14
df_winter.loc[df_winter['year'].isin(['2015'])] = df_wt15
df_winter.loc[df_winter['year'].isin(['2016'])] = df_wt16
df_winter.loc[df_winter['year'].isin(['2017'])] = df_wt17
df_winter.loc[df_winter['year'].isin(['2018'])] = df_wt18
df_winter.loc[df_winter['year'].isin(['2019'])] = df_wt19
df_winter.loc[df_winter['year'].isin(['2020'])] = df_wt20
```

```python
df_winter.isnull().sum()
```

```python
df_winter[['year','70T']]
```

```python
year_view = pd.DataFrame(df_winter.groupby(by='year')[['57T','73T','35T','36T','67T','75T','70T','70T**','69T','58T','37T','38T','39T','40T','68T']]
                              .mean())
year_view
```

```python
year_wt = year_view 
year_wt['mean'] = year_view.mean(axis=1)
year_wt[['mean']]
```

```python
dfw = year_wt
x = year_wt.index
y = year_wt[['mean']]
```
