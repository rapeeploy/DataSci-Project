```python
rain_frames = [df.loc['2011-05-16':'2011-10-15'],
                 df.loc['2012-05-16':'2012-10-15'],
                 df.loc['2013-05-16':'2013-10-15'],
                 df.loc['2014-05-16':'2014-10-15'],
                 df.loc['2015-05-16':'2015-10-15'],
                 df.loc['2016-05-16':'2016-10-15'],
                 df.loc['2017-05-16':'2017-10-15'],
                 df.loc['2018-05-16':'2018-10-15'],
                 df.loc['2019-05-16':'2019-10-15'],
                 df.loc['2020-05-16':'2020-10-15'],]

df_rain = pd.concat(rain_frames)
df_rain
```

```python
df_rain.isnull().sum()
```

```python
df_rain[df_rain.isnull().any(axis=1)] 

#แทนที่ด้วยค่าเฉลี่ยของแต่ละปี 
```

```python
df_rain11 = df_rain.loc['2011']
df_rain12 = df_rain.loc['2012']
df_rain13 = df_rain.loc['2013']
df_rain14 = df_rain.loc['2014']
df_rain15 = df_rain.loc['2015']
df_rain16 = df_rain.loc['2016']
df_rain17 = df_rain.loc['2017']
df_rain18 = df_rain.loc['2018']
df_rain19 = df_rain.loc['2019']
df_rain20 = df_rain.loc['2020']
```

```python
df_rain11 = df_rain11.fillna(df_rain11.mean())
df_rain12 = df_rain12.fillna(df_rain12.mean())
df_rain13 = df_rain13.fillna(df_rain13.mean())
df_rain14 = df_rain14.fillna(df_rain14.mean())
df_rain15 = df_rain15.fillna(df_rain15.mean())
df_rain16 = df_rain16.fillna(df_rain16.mean())
df_rain17 = df_rain17.fillna(df_rain17.mean())
df_rain18 = df_rain18.fillna(df_rain18.mean())
df_rain19 = df_rain19.fillna(df_rain19.mean())
df_rain20 = df_rain20.fillna(df_rain20.mean())
```

```python
df_rain.loc['2011'] = df_rain11
df_rain.loc['2012'] = df_rain12
df_rain.loc['2013'] = df_rain13
df_rain.loc['2014'] = df_rain14
df_rain.loc['2015'] = df_rain15
df_rain.loc['2016'] = df_rain16
df_rain.loc['2017'] = df_rain17
df_rain.loc['2018'] = df_rain18
df_rain.loc['2019'] = df_rain19
df_rain.loc['2020'] = df_rain20
df_rain
```

```python
df_rain['year'] = pd.DatetimeIndex(df_rain.index).year
df_rain
```

```python
year_view = pd.DataFrame(df_rain.groupby(by='year')[['57T','73T','35T','36T','67T','75T','70T','70T**','69T','58T','37T','38T','39T','40T','68T']]
                              .mean())
year_view
```

```python
year_ra = year_view 
year_ra['mean'] = year_view.mean(axis=1)
year_ra[['mean']]
```

```python
dfr = year_ra
x = year_ra.index
y = year_ra[['mean']]
```
